### **Imports**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from sklearn.model_selection import train_test_split
```

```python
sns.set_theme(palette='winter')
```

```python
train = pd.read_csv('datasets/clean-avg-train-combined-with-location-data.csv')
train.head()
```

```python
df=train[['month_beginning', 'station_name','month_total']].set_index('month_beginning')
df.head()
```

```python
df=df.groupby(df.index)[['month_total']].mean()
df.head()
```

```python
plt.figure(figsize=(18,9))
plt.plot(df['month_total'])
plt.xticks(df.index[0::1]);
plt.title('Average Monthly Ridership', fontsize=26)
plt.xlabel(None, fontsize=20)
```
